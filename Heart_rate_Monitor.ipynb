```python
chosen_set_f=[]
for i in range(len(chosen_set)):
    if chosen_set[i] >= threshold:
        chosen_set_f.append(chosen_set[i])
chosen_set_f  
```

```python
mean_set = np.mean(chosen_set_f)
print(mean_set)
max_diff = mean_set/4
print(max_diff)
R = []
for i in range(len(chosen_set_f)):
    if abs(chosen_set_f[i] - mean_set) > max_diff:
        R.append(chosen_set_f[i])
#print(R)
while R != []:
    for ele in R:
        chosen_set_f.remove(ele)
    R = []
    for i in range(len(chosen_set_f)):
        if abs(chosen_set_f[i] - mean_set) > max_diff:
            R.append(chosen_set_f)
print(chosen_set_f)
```
